```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

N = 11
data = np.random.uniform(-N, N+1, 100)
df = pd.DataFrame(data, columns=['Значения'])
print(df)

#A СТатистика
mean_value = df['Значения'].mean()
std_value = df['Значения'].std()
median_value = df['Значения'].median()
print(f"Среднее значение: {mean_value}")
print(f"Стандартное отклонение: {std_value}")
print(f"Медиана: {median_value}")

#Б (нечетный вариант) построение диаграммы размаха
plt.figure(figsize=(8,6))
plt.boxplot(df['Значения'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Ящичковая иаграмма размаха')
plt.xlabel('Значения')
plt.grid(True)
plt.show()
```
